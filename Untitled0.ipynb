```python
def survival_duration_calculator():
    print(" The Survival Duration Calculator!")

    # get user age #
    try:
        age = int(input("What's your age? "))
        if age < 0:
            print("Age cannot be negative. Please enter a valid age.")
            return
    except ValueError:
        print("Invalid input. Please enter a number.")
        return

    #  Duration in diff time units#
    print("Please choose a time unit: Months, Weeks, Days, Hours, Minutes, Seconds.")
    print("Note: You can write the first letter or the full name of the time unit.")
    time_unit = input("Your time unit: ").strip().lower()

    # Specify the conversions
    conversions =  {
        "months": 12,
        "m": 12,

        "weeks": 52,
        "w": 52,

        "days": 365,
        "d": 365,

        "hours": 365 * 24,
        "h": 365 * 24,

        "minutes": 365 * 24 * 60,
        "min": 365 * 24 * 60,

        "seconds": 365 * 24 * 60 * 60,
        "s": 365 * 24 * 60 * 60
    }

    # Calculation and display ofthe  selected time unit #
    if time_unit in conversions:
        duration = age * conversions[time_unit]
        unit_name = time_unit if len(time_unit) > 1 else [key for key, val in conversions.items() if val == conversions[time_unit] and len(key) > 1][0]
        print(f"You lived for {duration} {unit_name.capitalize()}.")

    else:
        print("Invalid time unit.")

# Run the calculator #
survival_duration_calculator()
```

```python
class SurvivalDurationCalculator:
    def __init__(self, age):
        """
        Initialize the calculator with the age.
        """
        self.age = age  # Age in years
        self.conversions = {
            "months": 12,
            "weeks": 52,
            "days": 365,
            "hours": 365 * 24,
            "minutes": 365 * 24 * 60,
            "seconds": 365 * 24 * 60 * 60,
        }

    def calculate_duration(self, time_unit):
        """
        Calculates the duration in the specified time unit.
        :param time_unit: The unit of time for which to calculate the duration.
        :return: The calculated duration.
        """
        time_unit = time_unit.strip().lower()
        if time_unit in self.conversions:
            return self.age * self.conversions[time_unit]
        else:
            raise ValueError("Invalid time unit selected.")

```
